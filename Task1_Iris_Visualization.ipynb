```python
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt

# Load the Iris dataset with pandas
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
iris_df = pd.read_csv(url)

# Inspect the dataset
print("Shape:", iris_df.shape)
print("Columns:", iris_df.columns.tolist())
print("\nFirst 5 rows:")
print(iris_df.head())

print("\nInfo:")
iris_df.info()

print("\nSummary statistics:")
print(iris_df.describe())

# Scatter plot: relationship between sepal length and sepal width
plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=iris_df,
    x="sepal_length",
    y="sepal_width",
    hue="species",
    style="species",
    s=80
)
plt.title("Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.tight_layout()
plt.show()

# Histograms: distributions of numeric features
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
numeric_cols = iris_df.select_dtypes(include="number").columns
for ax, col in zip(axes.flatten(), numeric_cols):
    sns.histplot(iris_df[col], kde=True, ax=ax, color="steelblue")
    ax.set_title(f"Histogram of {col}")
    ax.set_xlabel(col)
    ax.set_ylabel("Count")
plt.tight_layout()
plt.show()

# Box plots: identify outliers by species
plt.figure(figsize=(10, 6))
sns.boxplot(data=iris_df, x="species", y="sepal_length")
plt.title("Sepal Length Distribution by Species")
plt.xlabel("Species")
plt.ylabel("Sepal Length (cm)")
plt.tight_layout()
plt.show()
```
